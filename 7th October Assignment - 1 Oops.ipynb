Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
class Employee:
  def __init__(self,empID,name,salary):
    self.empID = empID
    self.name = name
    self.salary = salary

  def calculate_yearly_bonus(self, bonus_percentage):
      """
      Calculates the yearly bonus based on the given bonus percentage.
      Args:
          bonus_percentage (float): Bonus percentage (e.g., 10% as 0.1).

      Returns:
          float: Yearly bonus amount.
      """
      return self.salary * bonus_percentage

  def display_employee_details(self):
      """
      Displays employee details.
      """
      print(f"Employee ID: {self.empID}")
      print(f"Name: {self.name}")
      print(f"Salary: Rs{self.salary:.2f}")
```

Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}  # Dictionary to store available vehicles

    def rent_vehicle(self, vehicle_id):
        """
        Rent a vehicle by marking it as unavailable.
        Args:
            vehicle_id (str): Unique identifier for the vehicle.

        Returns:
            bool: True if successful, False if the vehicle is already rented or not found.
        """
        if vehicle_id in self.available_vehicles:
            self.available_vehicles[vehicle_id] = False
            return True
        else:
            print(f"Vehicle with ID {vehicle_id} not found or already rented.")
            return False

    def return_vehicle(self, vehicle_id):
        """
        Return a rented vehicle by marking it as available.
        Args:
            vehicle_id (str): Unique identifier for the vehicle.

        Returns:
            bool: True if successful, False if the vehicle is not found or already available.
        """
        if vehicle_id in self.available_vehicles:
            if not self.available_vehicles[vehicle_id]:
                self.available_vehicles[vehicle_id] = True
                return True
            else:
                print(f"Vehicle with ID {vehicle_id} is already available.")
                return False
        else:
            print(f"Vehicle with ID {vehicle_id} not found.")
            return False

    def display_available_vehicles(self):
        """
        Display details of available vehicles.
        """
        print("Available Vehicles:")
        for vehicle_id, is_available in self.available_vehicles.items():
            if is_available:
                print(f"Vehicle ID: {vehicle_id} (Available)")

# Example usage
if __name__ == "__main__":
    rental_system = VehicleRentalSystem()

    # Add some vehicles (vehicle_id: availability)
    rental_system.available_vehicles = {
        "V001": True,
        "V002": True,
        "V003": True,
    }

    rental_system.display_available_vehicles()

    # Rent a vehicle
    rental_system.rent_vehicle("V002")
    rental_system.display_available_vehicles()

    # Return a vehicle
    rental_system.return_vehicle("V001")
    rental_system.display_available_vehicles()
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Library:
    def __init__(self):
        self.books = {}  # Dictionary to store books {book_id: Book object}

    def add_book(self, book):
        """
        Adds a book to the library.
        Args:
            book (Book): Book object to add.
        """
        self.books[book.book_id] = book

    def borrow_book(self, book_id):
        """
        Borrows a book by marking it as unavailable.
        Args:
            book_id (str): Unique identifier for the book.

        Returns:
            bool: True if successful, False if the book is already borrowed or not found.
        """
        if book_id in self.books:
            if self.books[book_id].is_available:
                self.books[book_id].is_available = False
                return True
            else:
                print(f"Book with ID {book_id} is already borrowed.")
                return False
        else:
            print(f"Book with ID {book_id} not found.")
            return False

    def return_book(self, book_id):
        """
        Returns a borrowed book by marking it as available.
        Args:
            book_id (str): Unique identifier for the book.

        Returns:
            bool: True if successful, False if the book is not found or already available.
        """
        if book_id in self.books:
            if not self.books[book_id].is_available:
                self.books[book_id].is_available = True
                return True
            else:
                print(f"Book with ID {book_id} is already available.")
                return False
        else:
            print(f"Book with ID {book_id} not found.")
            return False

    def display_available_books(self):
        """
        Displays details of available books.
        """
        print("Available Books:")
        for book_id, book in self.books.items():
            if book.is_available:
                print(f"Book ID: {book_id} (Available)")

class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_available = True

# Example usage
if __name__ == "__main__":
    library = Library()

    # Add some books
    library.add_book(Book("B001", "The Great Gatsby", "F. Scott Fitzgerald"))
    library.add_book(Book("B002", "To Kill a Mockingbird", "Harper Lee"))
    library.add_book(Book("B003", "1984", "George Orwell"))

    library.display_available_books()

    # Borrow a book
    library.borrow_book("B002")
    library.display_available_books()

    # Return a book
    library.return_book("B001")
    library.display_available_books()
```

Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Product:
    def __init__(self, product_id, name, quantity):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity

class InventorySystem:
    def __init__(self):
        self.products = {}  # Dictionary to store products {product_id: Product object}

    def add_product(self, product):
        """
        Adds a product to the inventory.
        Args:
            product (Product): Product object to add.
        """
        self.products[product.product_id] = product

    def update_product_quantity(self, product_id, new_quantity):
        """
        Updates the quantity of a product in the inventory.
        Args:
            product_id (str): Unique identifier for the product.
            new_quantity (int): New quantity for the product.

        Returns:
            bool: True if successful, False if the product is not found.
        """
        if product_id in self.products:
            self.products[product_id].quantity = new_quantity
            return True
        else:
            print(f"Product with ID {product_id} not found.")
            return False

    def display_available_products(self):
        """
        Displays details of available products.
        """
        print("Available Products:")
        for product_id, product in self.products.items():
            print(f"Product ID: {product_id} (Name: {product.name}, Quantity: {product.quantity})")

# Example usage
if __name__ == "__main__":
    inventory_system = InventorySystem()

    # Add some products
    inventory_system.add_product(Product("P001", "Laptop", 10))
    inventory_system.add_product(Product("P002", "Smartphone", 5))
    inventory_system.add_product(Product("P003", "Headphones", 20))

    inventory_system.display_available_products()

    # Update product quantity
    inventory_system.update_product_quantity("P001", 15)
    inventory_system.display_available_products()
```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        """
        Calculates the area of the shape.
        """
        raise NotImplementedError("Subclasses must implement this method.")

    def calculate_perimeter(self):
        """
        Calculates the perimeter (or volume) of the shape.
        """
        raise NotImplementedError("Subclasses must implement this method.")


class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)


class Square(Rectangle):
    def __init__(self, side):
        super().__init__(length=side, width=side)


class Circle(Shape):
    def calculate_area(self):
        import math
        return math.pi * (self.length ** 2)

    def calculate_perimeter(self):
        import math
        return 2 * math.pi * self.length


class Triangle(Shape):
    def calculate_area(self):
        # Using Heron's formula for the area of a triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5

    def calculate_perimeter(self):
        return self.length + self.width + self.height


# Example usage
if __name__ == "__main__":
    # Create a rectangle
    rectangle = Rectangle(length=5, width=3)
    print(f"Rectangle area: {rectangle.calculate_area()}")
    print(f"Rectangle perimeter: {rectangle.calculate_perimeter()}")

    # Create a square
    square = Square(side=4)
    print(f"Square area: {square.calculate_area()}")
    print(f"Square perimeter: {square.calculate_perimeter()}")

    # Create a circle
    circle = Circle(length=2)
    print(f"Circle area: {circle.calculate_area()}")
    print(f"Circle circumference: {circle.calculate_perimeter()}")

    # Create a triangle
    triangle = Triangle(length=3, width=4, height=5)
    print(f"Triangle area: {triangle.calculate_area()}")
    print(f"Triangle perimeter: {triangle.calculate_perimeter()}")
```

Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.grades = []

    def add_grade(self, grade):
        """
        Add a grade to the student's record.
        """
        self.grades.append(grade)

    def calculate_average_grade(self):
        """
        Calculate the average grade for the student.
        """
        if not self.grades:
            return 0  # No grades available
        total = sum(self.grades)
        return total / len(self.grades)

    def display_details(self):
        """
        Display student details.
        """
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {', '.join(map(str, self.grades))}")
        print(f"Average Grade: {self.calculate_average_grade():.2f}")


# Example usage
if __name__ == "__main__":
    # Create a student
    student1 = Student(student_id="S12345", name="Alice")

    # Add grades
    student1.add_grade(85)
    student1.add_grade(92)
    student1.add_grade(78)

    # Display student details
    student1.display_details()
```

Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
class Email:
    def __init__(self, sender, recipient, subject):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = None

    def set_body(self, body):
        """
        Set the email body.
        """
        self.body = body

    def send(self):
        """
        Send the email.
        """
        # Code to send the email using an email service or library
        print(f"Email sent from {self.sender} to {self.recipient} with subject: {self.subject}")

    def display_details(self):
        """
        Display email details.
        """
        print(f"From: {self.sender}")
        print(f"To: {self.recipient}")
        print(f"Subject: {self.subject}")
        if self.body:
            print(f"Body: {self.body}")


# Example usage
if __name__ == "__main__":
    # Create an email
    email = Email(sender="sender@example.com", recipient="recipient@example.com", subject="Important Update")

    # Set the email body
    email.set_body("This is the body of the email.")

    # Send the email
    email.send()

    # Display email details
    email.display_details()
```

Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post):
        """
        Add a post to the profile.
        """
        self.posts.append(post)

    def display_posts(self):
        """
        Display all posts on the profile.
        """
        for post in self.posts:
            print(post)

    def search_posts(self, keyword):
        """
        Search posts by a given keyword.
        """
        results = []
        for post in self.posts:
            if keyword in post:
                results.append(post)
        return results


# Example usage
if __name__ == "__main__":
    # Create a social media profile
    profile = SocialMediaProfile("johndoe")

    # Add some posts
    profile.add_post("My first post!")
    profile.add_post("Check out this cool new place I found.")
    profile.add_post("Looking for recommendations for a good book to read.")

    # Display all posts
    profile.display_posts()

    # Search for posts containing the word "cool"
    results = profile.search_posts("cool")
    print(f"Found {len(results)} posts containing the word 'cool':")
    for post in results:
        print(post)
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date):
        """
        Add a task to the ToDo list with a due date.
        """
        self.tasks.append({"task": task, "due_date": due_date, "completed": False})

    def mark_task_as_completed(self, task):
        """
        Mark a task as completed.
        """
        for item in self.tasks:
            if item["task"] == task:
                item["completed"] = True
                break

    def display_pending_tasks(self):
        """
        Display all pending tasks.
        """
        for item in self.tasks:
            if not item["completed"]:
                print(f"- {item['task']} (Due: {item['due_date']})")


# Example usage
if __name__ == "__main__":
    # Create a ToDo list
    todo_list = ToDoList()

    # Add some tasks
    todo_list.add_task("Finish project report", "2023-04-15")
    todo_list.add_task("Buy groceries", "2023-04-17")
    todo_list.add_task("Call the doctor", "2023-04-20")

    # Mark a task as completed
    todo_list.mark_task_as_completed("Buy groceries")

    # Display pending tasks
    todo_list.display_pending_tasks()
```
